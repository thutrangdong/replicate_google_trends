```python
from pytrends.request import TrendReq
from datetime import datetime
import pandas as pd
import numpy as np
import matplotlib as plt
import requests
import json
import scipy.stats
%matplotlib inline

response = requests.get('https://api.coindesk.com/v1/bpi/historical/close.json?start=2011-05-01&end=2017-11-18')
json_data = json.loads(response.content)
bpi = pd.DataFrame.from_dict(json_data)
bpi.drop(['disclaimer', 'time'], axis=1, inplace=True)
bpi.drop(['updated', 'updatedISO'], inplace=True)
bpi['date'] = bpi.index
bpi['date'] = pd.to_datetime(bpi['date'], format="%Y-%m-%d")
bpi.index = range(2394)
cols = bpi.columns.tolist()
cols = cols[-1:] + cols[:-1]
bpi = bpi[cols] 
```

```python
google = pd.read_csv("data/google_trends.csv", ";")
datetime_object = datetime.strptime(google.loc[0][0], '%d.%m.%y')
google['date'] = pd.to_datetime(google['date'], format="%d.%m.%y")
google['search_volume'] = google['search_volume'].str.replace(',','.')
google['search_volume'] = google['search_volume'].apply(pd.to_numeric)
#google.set_index('date', inplace=True)
```

```python
a = scipy.stats.pearsonr(bpi.bpi, google.search_volume)
b = scipy.stats.pearsonr(bpi.bpi.iloc[0:2394], google.search_volume.iloc[0:2394])
print(a)
print(b)
```

```python
result = np.correlate(google.search_volume, bpi.bpi, 'full')
center = int(result.size/2)
variance = 10
result = result[center-variance:center+1+variance]

normalized_result = result / np.linalg.norm(result)
```

```python
limit = 10
y = normalized_result
y = [x**50 for x in y]
x = list(range(-limit,limit+1,1))
plt.pyplot.bar(x, y)
```

```python
data = pd.concat([bpi.bpi, google.search_volume], axis=1).dropna()
data.head()
```

```python
delta_t = 3
data['rolling_mean_search_volume']  = data['search_volume'].rolling(delta_t).mean().shift(1)
data['order'] = 0
data['order'][data.search_volume > data.rolling_mean_search_volume.shift(1)] = -1 # Sell
data['order'][data.search_volume < data.rolling_mean_search_volume.shift(1)] = 1 # Buy
data['order'].iloc[:delta_t] = 0
data['log_returns'] = data.order * np.log(data.bpi.shift(-1)) - data.order * np.log(data.bpi)
(np.exp(data.log_returns.cumsum()) - 1).plot();
```
